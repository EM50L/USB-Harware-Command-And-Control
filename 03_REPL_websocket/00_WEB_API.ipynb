## Acceso por http/wifi

Si esl dispositivo esta adecuadamente configurado*** se puede acceder a el via WEB.   
http://circuitpython.local    (MDNS)   
http://192.168.0.xx

Ojo para poder escribir tiene que estar desabilitado el modo "disco USB" / "USB storage"


### ***


Si se configuran los parametros CIRCUITPY_WIFI_SSID y CIRCUITPY_WIFI_PASSWORD en el archivo settings.toml.   

CircuitPython se conectar automticamente a la red Wi-Fi especificada al iniciar.

Tambien de puede configurar CIRCUITPY_WEB_API_PASSWORD, Para acceder al dispositivo via  web. (el acceso web solo se habilitar si la conexin Wi-Fi tiene xito al iniciar)

El servidor web est en el puerto 80 a menos que se configure otro puerto en  CIRCUITPY_WEB_API_PORT. 

Tambin se habilita MDNS. El nombre de la placa anunciado a la red se puede configurar con CIRCUITPY_WEB_INSTANCE_NAME.

Aqui tienes un ejemplo del contenido del archivo settings.toml   
Que se usara en las practicas


```toml
CIRCUITPY_WIFI_SSID = "Nombre_Red_Wifi_1"
CIRCUITPY_WIFI_PASSWORD = "Password_Red_Wifi_1"
CIRCUITPY_WEB_API_PASSWORD ="pwd" #Password de la interfaz web
CIRCUITPY_WEB_API_PORT = 80       #puerto de la interfaz
SSID   = "Nombre_Red_Wifi_2"
WPWD   = "Password_Red_Wifi_1"
bot_tk = "botXXXX:YYYY"   #token del bot telegram para el control remoto
```
