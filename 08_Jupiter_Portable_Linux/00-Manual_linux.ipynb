## Practicas Micropython - Jupyterlab 
**Ejecutando python en una placa/dispositivo IOT.**   



## Usaremos Jupyterlab Desktop 
- Se puede hacer portable
- No requiere permisos de administrador
- Entorno python portable en espacio de usuario

https://github.com/jupyterlab/jupyterlab-desktop/releases


1) Descomprimimos el paquete.   
```dpkg --extract JupyterLab-Setup-Debian.deb JupyterLab```



<div style="page-break-after: always"></div>

2) Arrancamos JupyterLab (es una app electron).   


<div style="page-break-after: always"></div>

3) Instalamos el entorno.   


- Es posible que tarde un rato.   
   

<div style="page-break-after: always"></div>

4) Abrimos el archivo **01-Instalar-Micropython.ipynt**
   

5) lo ejecutamos de nuevo.(asi se instalara el kernel para Micropython)   
       
   En la lista de kernels disponibles ya deberia aparecer micropython
   
   **Nota:** Los comandos ejecutados son:

```bash
pip install jupyter_micropython_kernel
python -m jupyter_micropython_kernel.install
jupyter kernelspec list
```

<div style="page-break-after: always"></div>

6) Hacemos lo mismo con **02-install-circuitpython.ipynb**  (asi se instala el kernel circuitpython)      
      
   En la lista de kernels disponibles ya deberia aparecer circuitpython

   **Nota:** Los comandos ejecutados son:

```bash
pip install "https://github.com/adafruit/circuitpython_kernel/archive/master.zip"
python -m circuitpython_kernel.install
jupyter kernelspec list
```

<div style="page-break-after: always"></div>

7) En el lanzador ya deberiamos ver los kernels.   
   

- Si abrimos una consola de Circuitpython y ejecutamos help()   
  Es posible que no detecte la placa. (por defecto solo detecta unas pocas placas)      


<div style="page-break-after: always"></div>

- Esto se soluciona facilmente editando board.py   
  y poniendo el VID de nuestra placa como se ve en la imagen.
  En mi caso la ruta es:
  ~/.config/jupyterlab-desktop/jlab_server/lib/python3.8/site-packages/circuitpython_kernel/board.py 
  

### Ya tenemos acceso ejecutar codigo en la placa.
